```sql
..show all the data I want to quiry
SELECT * 
FROM `workspace`.`default`.`bright_coffee_shop_analysis_case_study_1_1` 
limit 100;

...checking the first date where the data was recorded
SELECT MIN(transaction_date)
FROM bright_coffee_shop_analysis_case_study_1_1

...checking the last date where the data was recorded
SELECT MAX(transaction_date)
FROM bright_coffee_shop_analysis_case_study_1_1

...checking the date of which the stores starts to make sale and the time for last sale
SELECT MIN(transaction_time)
FROM bright_coffee_shop_analysis_case_study_1_1

SELECT MAX(transaction_time)
FROM bright_coffee_shop_analysis_case_study_1_1

...counting total number of transaction ID and qty
SELECT COUNT(transaction_id)
FROM bright_coffee_shop_analysis_case_study_1_1

SELECT COUNT(transaction_qty),
       CASE
       WHEN HOUR(transaction_time)>7 AND HOUR(transaction_time)<10 THEN 'MORNING RUSH'
       WHEN HOUR(transaction_time)>10 AND HOUR(transaction_time)<12 THEN 'BRUNCH'
       WHEN HOUR(transaction_time)>12 AND HOUR(transaction_time)<15 THEN 'DAYTIME'
       WHEN HOUR(transaction_time)>15 AND HOUR(transaction_time)<18 THEN 'AFTERNOON'
       ELSE 'EVENING'
       END AS time_intervals
FROM bright_coffee_shop_analysis_case_study_1_1
GROUP BY time_intervals;
            

..Checking names of where stores are located
SELECT DISTINCT store_location
FROM bright_coffee_shop_analysis_case_study_1_1

SELECT SUM(transaction_qty*unit_price),
       product_category
       FROM bright_coffee_shop_analysis_case_study_1_1
       GROUP BY product_category


```
